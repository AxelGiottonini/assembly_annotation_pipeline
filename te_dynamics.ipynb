## TE Dating

```python
files = glob.glob("./out/te_dynamics/*/dating/LTR_SNP_pos_summary")

i = 0
re_assembler = r"^./out/te_dynamics/(?P<assembler>.*)/dating/LTR_SNP_pos_summary$"
match = re.match(re_assembler, files[i]).groupdict()
assembler = match['assembler']

data = np.loadtxt(files[i], dtype=int).T

sns.lineplot(x=data[0], y=data[1], color="blue")
plt.title(f"Assembler: {assembler}")
plt.show()
```
